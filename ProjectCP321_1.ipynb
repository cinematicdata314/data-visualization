Our first representation is formed from our first dataset mentioned. It is using a Horizontal Bar graph to show the Average increase in housing prices between January 2021 - January 2022 in Canada by Province/Territory. We have organzied it in ascending order for the % Change in housing prices. 

From this visualization you can see why location is such an important factor when it comes to the the value of your house, depending on which province/territory you live in will determine weather your houses value will increase or decrease. Yukon and Northwest terrirores were the only ones that had a decease in housing value over the year which is due to the fact these provinces/territories have a very low population and most of the land inhabitable due to the cold. At the same time you can see that Ontario and BC are at the top of the list with the highest % increase in housing value over the year due to the increasing demand and population in these provinces. In conclusion, location is a key factor that contributes to weather a houses value will increase or not. 

```python
import pandas as pd
df = pd.read_csv('Average home prices in Canadian provinces.csv')
import matplotlib.pyplot as plt

# sort the DataFrame by % Change column in ascending order
df = df.sort_values('y/y % change')

#Exclude Canada from data
df = df[df['Area'] != 'Canada']

# create the horizontal bar chart
plt.barh(df['Area'], df['y/y % change'],color=['red' if x < 0 else 'green' for x in df['y/y % change']])

plt.title('Average Percentage Increase in Housing Prices by Province/Territory (Jan 2021 - Jan 2022)')
plt.xlabel('% Change in Housing Prices')
plt.ylabel('Province/Territory')

# display the chart
plt.show()
```

Our second representation is formed from the second data set mentioned. It is using a line graph to show the Average house hold income compared to the Median House value by city in California. The graph is set up so that the first three cities, Los Angeles, San Diago, and San Francisco, are one of the most highly populated cities in California while the last three cities, Siskiyou, Trinity and Tuolumne, have significatly low populations. The reason for this is to show the importance of population and how it effects the value of a house. 

From this visualization you can see that for the highly populated cities, the Median house value (represented by the orange line) is significantly higher than that of the cities with low populations. The median house value for the highly populated cities is easily above $600,000 while for the low populated cities the average ranges around $200,000 - $300,000. At the same time you can see that the average household income (represented by the blue line) for the highly populated cities is higher than that of the lower populated cites. Therefore, from this visualization, you can see that the household income and popluation density are key factors that contribute to a houses value. Cities, with a higher population density result in a higher household income, which therefore results in a higher housing value, which will keep on increasing overtime due to the increase in popularity and population of the cities. 

```python
import pandas as pd
import matplotlib.pyplot as plt

# read the CSV file into a pandas DataFrame
df = pd.read_csv('california_housing_updated.csv')

# choose the cities you want to graph
selected_cities = ['Los Angeles County', 'San Diego County', 'San Francisco County', 'Tuolumne County', 'Trinity County', 'Siskiyou County']

# filter the DataFrame to include only the rows for the selected cities
selected_data = df[df['name'].isin(selected_cities)]

# group the selected data by city name and calculate the average household income and median house value for each city
city_data = selected_data.groupby('name')[['average_household_income', 'median_house_value']].mean()

# set the tick labels on the x-axis to be the city names and rotate them by 90 degrees
plt.xticks(rotation=90)

# plot a line graph of the average household income for each city
plt.plot(city_data.index, city_data['average_household_income'], label='Average Household Income')

# plot a line graph of the median house value for each city in a different color
plt.plot(city_data.index, city_data['median_house_value'], label='Median House Value')

plt.xlabel('City')
plt.ylabel('Dollars')
plt.title('Average Household Income and Median House Value by City in California')

# add a legend to the plot
plt.legend()

plt.show()
```
