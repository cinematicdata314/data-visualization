Our dataset contains weekly retail prices of three grades of gasoline, diesel, auto propane and compressed natural gas in 10 Ontario markets. It is downloaded from https://data.ontario.ca/dataset/fuels-price-survey-information. Please plot the following visualizations. Your visualization should be complete.

You can feel free to use the plotting packages in python. Please submit the ipynb file only.



```python
import pandas as pd
df = pd.read_csv('fueltypesall1991-2023.csv')
```

## Question 1

Compare the mean Regular Unleaded Gasoline price among different cities in 2022. Plot the corresponding bar chart.

```python
#df.head()
#df.shape
import matplotlib.pyplot as plt

total = df[(df['Fuel Type'] == 'Regular Unleaded Gasoline') & (df['Date'].str.contains('2022'))].mean()
#print(str(total))

total.plot(kind='bar')
plt.title("Mean Regular Unleaded Gasoline price among different cities in 2022")
plt.xlabel("Mean Price")
plt.ylabel("Cities")
plt.show()
```

## Question 2

Plot the histogram of the Regular Unleaded Gasoline price in Ottawa for 2022. Choose the appropriate parameters.

```python
import matplotlib.pyplot as plt

total = df['Ottawa'][(df['Fuel Type'] == 'Regular Unleaded Gasoline') & (df['Date'].str.contains('2022'))]
#print(str(total))

total.plot(kind='hist', alpha = 0.5)
plt.title("Regular Unleaded Gasoline price in Ottawa for 2022")
plt.xlabel("Price")
plt.ylabel("Density")
plt.show()
```

## Question 3

Plot the ECDF (normalized) of the Regular Unleaded Gasoline price in Ottawa for 2022.

```python
import seaborn as sns
import matplotlib.pyplot as plt

total = df['Ottawa'][(df['Fuel Type'] == 'Regular Unleaded Gasoline') & (df['Date'].str.contains('2022'))]
#print(str(total))

sns.ecdfplot(x = total)
plt.title("ECDF (normalized) of the Regular Unleaded Gasoline price in Ottawa for 2022")
plt.xlabel('Price')
plt.ylabel('ECDF Proportion')
plt.show()
```
