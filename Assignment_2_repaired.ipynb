Visualize the time series: Regular Unleaded Gasoline price in "London" from 2000 to 2022. Smooth the time series using simple moving average with windowsize=5.

```python
# === Q1 starter (you can replace/modify as needed) ===
import pandas as pd
import matplotlib.pyplot as plt

# 1) Load your dataset (ensure the CSV is in the same folder as the notebook)
#    If the filename differs, change it here.
csv_path = 'fueltypesall1991-2023.csv'
df = pd.read_csv(csv_path)

print('Columns in your dataset:', list(df.columns))

# 2) Adjust these column names to match your file
DATE_COL = 'Date'              # e.g., 'Week of' or 'Date'
CITY_COL = 'Market'            # e.g., 'City' or 'Market'
PRODUCT_COL = 'Fuel Type'      # e.g., 'Product' or 'Fuel Type'
PRICE_COL = 'Price'            # e.g., 'Price (cents/litre)' or similar

# 3) Filter: London & Regular Unleaded Gasoline
mask = (df[CITY_COL].astype(str).str.strip().str.lower() == 'london') &        (df[PRODUCT_COL].astype(str).str.contains('regular', case=False, na=False))
sub = df.loc[mask].copy()

# 4) Parse dates and filter to 2000–2022 inclusive
sub[DATE_COL] = pd.to_datetime(sub[DATE_COL], errors='coerce')
sub = sub.dropna(subset=[DATE_COL])
sub = sub[(sub[DATE_COL].dt.year >= 2000) & (sub[DATE_COL].dt.year <= 2022)]
sub = sub.sort_values(DATE_COL)

# 5) Optional: if price is in cents/litre, convert to dollars/litre for readability
y = PRICE_COL
y_label = 'Price'
if sub[PRICE_COL].max() > 50:  # crude heuristic — adjust as needed
    sub['price_dollars'] = sub[PRICE_COL] / 100.0
    y = 'price_dollars'
    y_label = 'Price ($/L)'

# 6) Simple moving average with window=5
sub['SMA_5'] = sub[y].rolling(window=5, min_periods=1).mean()

# 7) Plot
plt.figure(figsize=(10, 5))
plt.plot(sub[DATE_COL], sub[y], label='Weekly Price')
plt.plot(sub[DATE_COL], sub['SMA_5'], label='5-Week SMA')
plt.title('Regular Unleaded Gasoline – London (2000–2022)')
plt.xlabel('Date')
plt.ylabel(y_label)
plt.legend()
plt.tight_layout()
plt.show()
```
